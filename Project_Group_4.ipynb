Dropping nulls as we can still retain 80% of the data which is huge enough

```python
df2 = df1.dropna()
```

```python
df2.shape
```

```python
df2.drop_duplicates(inplace=True)
```

```python
# df.fillna(df.mean(), inplace=True)
# df.fillna(df.mode().iloc[0], inplace=True)
```

```python
df2.isna().sum()
```

```python
df2['Start_Time'] = pd.to_datetime(df2['Start_Time'])
df2['End_Time'] = pd.to_datetime(df2['End_Time'])
```

```python
# Add column for day of week
df2['Day_of_Week'] = df2['Start_Time'].dt.day_name()

# Add column for hour of day
df2['Hour_of_Day'] = df2['Start_Time'].dt.hour

# Add column for month of the accident incident
df2['Month'] = df2['Start_Time'].dt.month

# Add column for year of the accident inciden
df2['Year'] = df2['Start_Time'].dt.year

# Add column for duration of accident (in minutes)
df2['Duration'] = (df2['End_Time'] - df2['Start_Time']).dt.total_seconds() / 60
```

```python
df2.columns
```

```python
df3 = df2[['ID', 'Severity', 'Start_Lat', 'Start_Lng',
       'End_Lat', 'End_Lng','Hour_of_Day', 'Duration','Day_of_Week', 'Month', 'Year',
       'Distance(mi)', 'City', 'County', 'State',
       'Zipcode', 'Country', 'Timezone', 'Temperature(F)', 'Wind_Chill(F)',
       'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)',
       'Precipitation(in)', 'Weather_Condition', 'Bump', 'No_Exit', 'Railway',
       'Roundabout', 'Stop', 'Traffic_Signal', 'Sunrise_Sunset']]
```

```python
df3
```

```python
# Categorize Severity column
df3['Severity'] = df3['Severity'].replace({1: 'Minor', 2: 'Moderate', 3: 'Severe', 4: 'Very Severe'})
```

```python
# Count the number of accidents in each state
state_counts = df3['State'].value_counts()

# Create a bar chart
plt.figure(figsize=(18, 10))
plt.bar(state_counts.index, state_counts.values)
plt.xlabel('State')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
# Create a heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(df3.corr())
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(15, 12))
plt.scatter(df3['Hour_of_Day'], df3['Severity'])
plt.xlabel('Hour_of_Day')
plt.ylabel('Severity')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count accidents by severity
severity_counts = df3['Severity'].value_counts()

# Create bar chart
plt.figure(figsize=(10, 8))
plt.bar(severity_counts.index, severity_counts.values)
plt.title('Accidents by Severity')
plt.xlabel('Severity')
plt.ylabel('Count')
plt.show()
```

```python
# Count accidents by hour of day
hour_counts = df3.groupby('Hour_of_Day')['ID'].count()

# Create line plot
plt.figure(figsize=(10, 8))
plt.plot(hour_counts.index, hour_counts.values)
plt.title('Accidents by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns

# Count accidents by day of week and hour of day
day_hour_counts = df3.pivot_table(index='Day_of_Week', columns='Hour_of_Day', values='ID', aggfunc='count')

# Create heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(day_hour_counts, cmap='YlGnBu')
plt.title('Accidents by Day of Week and Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.show()
```
